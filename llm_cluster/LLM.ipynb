```python
import numpy as np
import pandas as pd
from langchain_ollama import OllamaLLM
from langchain_core.prompts import PromptTemplate
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
```

```python
prompt_examples = ["You are an astrophysicist, explain how blackholes cause time dilation",
                 "You are an aerospace engineer, explain why the industry is moving towards reusable rockets"]
```

```python
import os
print(os.environ.get("OLLAMA_HOST")) 

# os.environ["OLLAMA_HOST"] = "http://localhost:11400"
```

```python
llm = OllamaLLM(model="qwen2.5:3b")

for prompt in prompt_examples:
    print(f"Evaluating prompting style: {prompt}")

    response = llm.invoke(prompt)

    print(response)
    print("-" * 80)
```
